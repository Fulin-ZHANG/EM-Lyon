```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have a DataFrame df_p4 with columns 'Weight(kg)' and 'Height(cm)'
# and the index contains the sport names

x = df_p4['Weight(kg)']
y = df_p4['Height(cm)']

# Create a scatter plot for non-outliers
plt.scatter(x, y, color='b', label='Data Points')

# Fit a linear regression line
linear_model = np.polyfit(x, y, 1)
linear_model_fn = np.poly1d(linear_model)
x_s = np.arange(40, 100)
plt.plot(x_s, linear_model_fn(x_s), color="r", label='Linear Fit Line')

# Define a threshold for identifying outliers
outlier_threshold = 3.0

# Annotate and change the color of outliers
legend_flag = False  # Initialize the flag
for sport, x_val, y_val in zip(df_p4.index, x, y):
    if np.abs(linear_model_fn(x_val) - y_val) > outlier_threshold:
        if not legend_flag:  # Check the flag before adding to the legend
            plt.scatter(x_val, y_val, color='red', label='Outliers')
            legend_flag = True  # Set the flag to True after adding to the legend
        else:
            plt.scatter(x_val, y_val, color='red')
        plt.annotate(sport, (x_val, y_val), textcoords='offset points', xytext=(0, 10),
                     ha='center', fontsize=8, color='red')

# Customize the plot
plt.title("Scatter Plot of the Weight and Height Average")
plt.xlabel("Avg-weight")
plt.ylabel("Avg-height")
plt.legend()

# Display the plot
plt.show()
```

__Question 2. (bonus)__
Add a color scale representing the number of data we have for each point.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

# Assuming you have a DataFrame df_p4 with columns 'Weight(kg)' and 'Height(cm)'
# and the index contains the sport names

x = df_p4['Weight(kg)']
y = df_p4['Height(cm)']

# Calculate the data density using a 2D kernel density estimation
kde = gaussian_kde([x, y])
density = kde([x, y])

# Create a colormap for the color scale
cmap = plt.get_cmap('viridis')

# Normalize the density values to fit the colormap
norm = plt.Normalize(vmin=density.min(), vmax=density.max())

# Create a scatter plot with colors representing data density
plt.scatter(x, y, c=density, cmap=cmap, s=50, label='Data Points', alpha=0.7)

# Fit a linear regression line
linear_model = np.polyfit(x, y, 1)
linear_model_fn = np.poly1d(linear_model)
x_s = np.arange(40, 100)
plt.plot(x_s, linear_model_fn(x_s), color="r", label='Linear Fit Line')

# Define a threshold for identifying outliers
outlier_threshold = 3.0

# Annotate and change the color of outliers
legend_flag = False  # Initialize the flag
for sport, x_val, y_val, dens in zip(df_p4.index, x, y, density):
    if np.abs(linear_model_fn(x_val) - y_val) > outlier_threshold:
        if not legend_flag:  # Check the flag before adding to the legend
            plt.scatter(x_val, y_val, color='red', label='Outliers')
            legend_flag = True  # Set the flag to True after adding to the legend
        else:
            plt.scatter(x_val, y_val, color='red')
        plt.annotate(sport, (x_val, y_val), textcoords='offset points', xytext=(0, 10),
                     ha='center', fontsize=8, color='red')

# Create a colorbar for the color scale
cbar = plt.colorbar(orientation='vertical', pad=0.04)
cbar.set_label('Data Density')

# Customize the plot
plt.title("Scatter Plot of the Weight and Height Average")
plt.xlabel("Avg-weight")
plt.ylabel("Avg-height")
plt.legend()

# Display the plot
plt.show()
```
