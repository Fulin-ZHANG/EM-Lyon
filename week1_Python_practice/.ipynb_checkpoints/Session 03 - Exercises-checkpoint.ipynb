### OOP

<div style=' background-color:#faf9e8'> 
    <b>Q1</b> Modeling with OOP.
</div>

Write a Python program to create a class representing a shopping cart. Include methods for adding and removing items, and calculating the total price.  

```python
class cart:
    def __init__(self,name):
        self.name = name
        self.items = {}
        
    def add_items(self, new_item, new_price):
        self.items[new_item] = new_price
            
    def remove_items(self,item_remove):
        del self.items[item_remove]
        
    def total_price(self):
        print(sum(self.items.values()))
            
    def __str__(self):
        return f"This is the {self.name} cart."
```

```python
cart1 = cart('Cart1')
```

```python
print(cart1)
```

```python
cart1.add_items('apple',100)
```

```python
cart1.remove_items('apple')
```

```python
cart1.total_price()
```

### Files

__Q2__ Even Numbers.

You are given a __.txt__ file called "evennumbers.txt".

Write a script that reads in the file and prints the sum of all the even numbers in this file.

```python
with open('./data/evennumbers.txt', mode = 'r') as myfile:
    out = myfile.readlines()
    result = 0
    for text in out:
        text.replace('\n','')
        if int(text)%2 == 0:
            result += int(text)
    print(out)
    print(result)
    myfile.close()
```

You are given a .txt file called "mbox-short.txt".

Create a function that reads in the file and looks for lines with "http" in them. (Hint: "http" can appear inside a longer string like "https".) Keep a count of how many URLs appear. Print out the first URL in each line, but nothing else in that line. This means you need to isolate the URL from the rest of the line (you can use space as the delimiter.)

At the end, return the total count of lines containing http in the file.

```python
with open('./data/mbox-short.txt', 'r') as my_file:
    out = my_file.readlines()
    for line in out:
        print(line)
    my_file.close()
```

```python
with open('./data/mbox-short.txt', 'r') as my_file:
    out = my_file.readlines()
    count_http = 0
    for line in out:
        if "http" in line:
            count_http += 1
            pos1 = line.find("http")
            line_cut = line[pos1:]
            print(line_cut)
    my_file.close()
```

```python
with open('./data/mbox-short.txt', 'r') as my_file:
    out = my_file.readlines()
    count_http = 0
    for line in out:
        if "http" in line:
            count_http += 1
            pos1 = line.find("http")
            line_cut = line[pos1:]
            if line_cut.find(")") == -1:
                if line_cut.find(" ") != -1:
                    pos2 = pos1 + line_cut.find(" ")
                    print (line[pos1:pos2])
                print(line[pos1:])
            else:
                pos2 = pos1 + line_cut.find(")")
                print(line[pos1:pos2])
    print(f"There are {count_http} lines")
    my_file.close()
```

<div style=' background-color:#faf9e8'> 
    <b>Q4</b> Normal Distribution.
</div>

Write a function that computes the normal distribution function with parameters `x`, `mu` ($\mu$) and `sigma` ($\sigma$) as:

<div>
<img style="height:100px" src="images/normal.png" />
</div>

```python
import math
def normal_pdf(x, mu, sigma):
    out = (1/(sigma * math.sqrt(2*math.pi))) * math.exp(-(pow((x-mu)/sigma,2))/2)
    
    return out
```

```python
import numpy as np
def normal_pdf(x, mu, sigma):
    out = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu)**2) / (2 * sigma**2))
    
    return out
```

2- Test your function with `x` being 1, 0 and -1. Set `mu=0` and `sigma=1`.

```python
print(normal_pdf(1,0,1))   # The expected output is: 0.24
print(normal_pdf(0,0,1))   # The expected output is: 0.39
print(normal_pdf(-1,0,1))  # The expected output is: 0.24
```

3- Create a 2D plot for `x` values in the range of -5 and 5 against the `normal_pdf(x)` values.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 11)

mu = 0
sigma = 1

y = normal_pdf(x,mu,sigma)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Normal PDF', color='black')
plt.scatter(x, y, color = 'red')
plt.scatter(1.5, normal_pdf(1.5, mu, sigma),label = 'Value at x = 1.5' ,color = 'b')
plt.title('Normal Distribution PDF')
plt.xlabel('x')
plt.ylabel('PDF')
# plt.grid(True)
plt.legend()

plt.axvline(1.5, color = 'Aqua', linestyle = 'dotted')
plt.axhline(normal_pdf(1.5, mu, sigma), color = 'Aqua', linestyle = 'dotted')


plt.show()
```
